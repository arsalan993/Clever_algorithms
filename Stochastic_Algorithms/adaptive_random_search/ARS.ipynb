```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Fri Nov  1 11:56:59 2019

@author: arsalan/
"""
import numpy as np 
import matplotlib.pyplot as plt 
from random import random, uniform
import pandas as pd
import keras
import sklearn
```

```python
from keras.models import load_model
model = load_model('regression.h5')
```

```python
test_bench = pd.read_csv("test_data.csv",header=None)
```

```python
x_ = test_bench.iloc[:,0:2].to_numpy()
x_ = np.expand_dims(x_,2)
y_ = test_bench.iloc[:,2].to_numpy()
```

```python
def objective_function(x_input):
    x_input = np.expand_dims(np.expand_dims(np.array(x_input),0),2)
    out = model.predict(x_input)
    return out[0][0]

def rand_in_bounds(min_value, max_value):
    x1_ = uniform(min(min_value), max(min_value))
    x2_ = uniform(min(max_value), max(max_value))
    return [x1_,x2_]


def random_vector(min_max):
    x1 = min_max[0]
    x2 = min_max[1]
    return rand_in_bounds(x1,x2)

def distance(targ,pred):
    return abs(targ-pred)
```

```python
def take_step(minmax, current, step_size):
    arr = []
    print(len(current))
    for i in range(len(current)):
        _min = max([minmax[i][0], current[i] - step_size])
        _max = min([minmax[i][1], current[i] + step_size])
        arr.append(uniform(_min, _max))
    return arr
```

```python
def large_step_size(iter_, step_size,s_factor,l_factor,iter_mult):
    if iter_ % iter_mult == 0:
        return step_size * l_factor
    return step_size * s_factor
```

```python
def take_steps(bounds, current, step_size, big_stepsize):
    step = {}
    big_step = {}
    step['vector'] = take_step(bounds,current['vector'],step_size)
    step['cost'] = objective_function(step['vector'])
    big_step['vector'] = take_step(bounds,current['vector'],big_stepsize)
    big_step['cost'] = objective_function(big_step['vector'])
    return step, big_step
```

```python
def distance(targ,pred):
    return abs(targ-pred)
```

```python
def search(max_iter,bounds,init_factor,s_factor,l_factor,iter_mult,max_no_impr,target):
    step_size_1 = (bounds[0][1] - bounds[0][0]) * init_factor
    step_size_2 = (bounds[1][1] - bounds[1][0]) * init_factor
    current = {}
    count = 0
    current['vector'] = random_vector(bounds)
    current['cost']   = objective_function(current['vector'])
    for iter_ in range(0,max_iter):
        big_stepsize = large_step_size(iter_,step_size_1,s_factor,l_factor,iter_mult)
        step, big_step = take_steps(bounds, current, step_size_1, big_stepsize)
        if distance(target,step['cost']) <= distance(target,current['cost']) or distance(target,big_step['cost']) <= distance(target,current['cost']):
            if distance(target,big_step['cost']) <= distance(target,step['cost']):
                step_size,current = big_stepsize , big_step
            else:
                current = step
            count = 0
        else:
            count+=1
            if count >= max_no_impr:
                count = 0
                step_size = step_size/s_factor
        print(' > iteration=%d, best=%f' % (iter_ + 1, current['cost']))
    return current
```

```python
n = 700
x_test = x_[n:n+1,:,:]
y_test = y_[n]
x_test,y_test
def main():
    #
    problem_size = 2
    bounds = [[0,1]] * problem_size
    #
    max_iter = 5000
    init_factor = 0.081
    s_factor = 1.2
    l_factor = 3.2
    iter_mult = 112
    max_no_impr = 32
    #
    best = search(max_iter,bounds,init_factor,s_factor,l_factor,iter_mult,max_no_impr,y_test)
    print('Done. Best Solution: model output=%f, model input=%s' % (best['cost'], str(best['vector'])))


main()
```
